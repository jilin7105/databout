```python
import pandas  as pd 
df = pd.read_csv("data/HR.csv")
df.head(10)

```

```python

'''
集中趋势
均值 ，中位数，众数，分位数

四分位数计算方法
Q1 = （n+1）*0.25  下四分位数
Q2 = （n+1）*0.5   
Q3 = （n+1）*0.75

离中趋势
标准差，方差 
方差  =  （（每个数字 - 数字的均值）相加 ）^2 /数字数量
标准差 的平方  = 方差

标准差 或者 方差： 越大表示数据越离散 越小表示数据越聚拢


正态分布 ：
 均值 u     
 数据在  u +- 1 *标准差  附近的概率 为 0.69
 数据在  u +- 1.96 *标准差  附近的概率 为 0.95
 数据在  u +- 2.58 *标准差  附近的概率 为 0.99

'''
```
